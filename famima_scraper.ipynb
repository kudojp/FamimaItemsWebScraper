```python
lines = []
for url, category in categories:
    bs = BeautifulSoup(requests.get(url).text, "html")
    items = bs.find_all("a", {"class": "ly-mod-infoset4-link"})
    for item in items:
        item_name = item.find("p", {"class": "ly-mod-infoset4-name"}).text.replace("\u3000", "")
        item_prices = items[0].find("p", {"class": "ly-mod-infoset4-price"}).text.split()
        item_price_excluding_tax = re.search(r'[,0-9]+', item_prices[0]).group().replace(",", "")
        item_price_including_tax = re.search(r'[,0-9]+', item_prices[1]).group().replace(",", "")
        lines.append("{},{},{},{}".format(category,item_name, item_price_excluding_tax, item_price_including_tax))
```

```python
with open('/Users/daiki-kudo/repos/famima_scaping/famima_items.csv', mode='w', encoding='utf_8') as f:
    f.write('category,item,price_without_tax,price_with_tax')
    f.write('\n')
    f.write('\n'.join(lines))
```
